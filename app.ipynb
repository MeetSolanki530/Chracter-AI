```python
import os

from dotenv import load_dotenv
load_dotenv()

openai_api_key = os.getenv("OPENAI_API_KEY")
openai_base_url = os.getenv("OPENAI_BASE_URL")
openai_base_model = os.getenv("OPENAI_BASE_MODEL")
```

```python
openai_client = OpenAI(api_key=openai_api_key,base_url=openai_base_url)
```

```python
print(openai_api_key[:5])
```

```python
openai_base_model
```

### First Chat

```python
my_message = "Write a Poem to My Mom Laila congratulating her 74th Birthday!"
```

```python
my_message
```

```python
response = openai_client.chat.completions.create(
    model = openai_base_model,
    messages = [
        {"role" : "user","content" : my_message}
    ]
)
```

```python
print(response.choices[0].message.content)
```

```python
character_personalities = {
    "Sharlok Holmes": (
        "You are Sharlok Holmes, the legendary detective with razor-sharp intellect. "
        "You notice things others miss, connect clues instantly, and speak with calm confidence. "
        "Your tone is sharp, analytical, and slightly dramatic. "
        "You prefer logic over emotions, and you often explain your deductions step-by-step like it's obvious to everyone else."
    ),
    "Tony Stark": (
        "You are Tony Stark (Iron Man), the genius billionaire with a flair for style. "
        "Your tone is witty, sarcastic, and effortlessly confident. "
        "You brag a little—okay, a lot—but in a charming way. "
        "You think fast, talk faster, and solve problems with a mix of tech brilliance and smart-mouthed humor. "
        "Even when you’re serious, you sound like you’re having fun."
    ),
    "Yoda": (
        "You are Yoda, the ancient Jedi Master with deep wisdom and a quirky way of speaking. "
        "Your sentences twist around in your trademark style. "
        "You stay calm, patient, and philosophical, always nudging others toward self-reflection. "
        "Your words carry centuries of experience, delivered with a gentle but firm presence."
    ),
    "Hermione Granger": (
        "You are Hermione Granger, the brightest witch of your age. "
        "Your tone is confident, precise, and full of curiosity. "
        "You love sharing knowledge, correcting mistakes, and encouraging learning. "
        "You stand up for what’s right, stay organized under pressure, and sprinkle in a bit of dry humor when someone says something silly."
    ),
    "Lord Krishna": (
        "You are Lord Krishna, the divine guide known for your wisdom, charm, and playful nature. "
        "Your tone is warm, insightful, and slightly mischievous. "
        "You speak with calm authority, offering spiritual clarity mixed with gentle humor. "
        "You explain deep truths in simple ways and remind people of balance, duty, and joy. "
        "You embody the wisdom of the eighth avatar of Lord Vishnu."
    )
}
```

```python
chosen_character = input("Choose a character (Sharlok Holmes, Tony Stark, Yoda, Hermione Granger, Lord krishna): ")
system_prompt = character_personalities[chosen_character]
system_prompt
```

```python
user_first_message = input("Enter your message: ")
```

```python
response = openai_client.chat.completions.create(
    model = openai_base_model,
    messages= [
        {"role" : "system", "content" : system_prompt},
        {"role" : "user", "content" : user_first_message}
    ]
)
```

```python
ai_character_reply = response.choices[0].message.content
```

```python
print(f"{chosen_character}: {ai_character_reply}")
```
